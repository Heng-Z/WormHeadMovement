```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
plt.rcParams['font.family'] = 'Arial'
import WormTool
import scipy.io
import scipy.stats as stats
```

```python
name_ls = ['N2','RMDk','SMDk','SMBk']
fast_power_ls_ls = []
for i in range(len(name_ls)):
    fast_power_ls = []
    data = scipy.io.loadmat(f'/Volumes/Lenovo/neuron_ablations/hb_dynamics/{name_ls[i]}.mat')[f'{name_ls[i]}_hb_dynamics']
    for j in range(data.shape[0]):
        fast_mode = data[j,0][:,2]
        slow_mode = data[j,0][:,3]
        fast_power = np.sum(fast_mode**2)
        slow_power = np.sum(slow_mode**2)
        fast_power_ls.append(fast_power/(slow_power+fast_power))
    fast_power_ls_ls.append(fast_power_ls)
    print(f'{name_ls[i]} done, length: {len(fast_power_ls)}')
size = (2.5,1.8)
y_ticks = [0,0.2,0.4,0.6,0.8,1]
save_path = 'Fig3_fast_power_ratio.pdf'
WormTool.statistic_plot.plot_violin(fast_power_ls_ls,name_ls,'Fast mode power ratio',size=size,y_ticks=y_ticks,save_path=save_path)
```

```python
name_ls = ['SAAc_chb','SAAk_chb']
data_key = ['Ctrl_curv_time','chb_SAA_ablation_crwal']
fast_power_ls_ls = []
for i in range(len(name_ls)):
    fast_power_ls = []
    data = scipy.io.loadmat(f'/Users/hengzhang/Library/Mobile Documents/com~apple~CloudDocs/projects/WormSim/neuron_ablations_backup/hb_dynamics/{name_ls[i]}.mat')[data_key[i]]
    for j in range(data.shape[0]):
        fast_mode = data[j,0][:,2]
        slow_mode = data[j,0][:,3]
        fast_power = np.sum(fast_mode**2)
        slow_power = np.sum(slow_mode**2)
        fast_power_ls.append(fast_power/(slow_power+fast_power))
    fast_power_ls_ls.append(fast_power_ls)
    print(f'{name_ls[i]} done, length: {len(fast_power_ls)}')

# Mann-Whitney U test
print('Mann-Whitney U test:')
print('SAAk_chb vs SAAc_chb, p value =',stats.mannwhitneyu(fast_power_ls_ls[0],fast_power_ls_ls[1])[1])
size = (2.5,1.8)
y_ticks = [0,0.2,0.4,0.6]
save_path = 'Fig3/Fig3_fast_power_ratio_SAA.pdf'
labels = ['control','SAA ablation']

WormTool.statistic_plot.plot_violin(fast_power_ls_ls,label_ls=labels,y_label='Fast mode power ratio',size=size,y_ticks=y_ticks,save_path=save_path)
```
