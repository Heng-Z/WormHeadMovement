```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
plt.rcParams['font.family'] = 'Arial'
import WormTool
import scipy.io
import scipy.stats as stats
```

```python
name_ls = ['RMEi_ctrl','RMEi']
data_key = ['RMEi_ctrl_hb_dynamics','RMEi_hb_dynamics']
amp_ls_ls = []
for i in range(len(name_ls)):
    amp_ls = []
    data = scipy.io.loadmat(f'/Users/hengzhang/Library/Mobile Documents/com~apple~CloudDocs/projects/WormSim/neuron_ablations_backup/hb_dynamics/{name_ls[i]}.mat')[data_key[i]]
    for j in range(data.shape[0]):
        curvature = data[j,0][:,1]
        amp_ls.append(WormTool.timeseries.calc_amplitude(curvature))
    amp_ls_ls.append(amp_ls)
    print(f'{name_ls[i]} done, length: {len(amp_ls)}')

# Mann-Whitney U test
print('Mann-Whitney U test:')
print('p value =',stats.mannwhitneyu(amp_ls_ls[0],amp_ls_ls[1])[1])
size = (2.5,1.8)
y_ticks = [2,4,6,8,10,12]
save_path = 'Fig3/Fig3_curvature_amplitude_RME.pdf'
labels = ['control','RME inhibition']

WormTool.statistic_plot.plot_violin(amp_ls_ls,label_ls=labels,y_label='Head Amplitude',size=size,y_ticks=y_ticks,save_path=save_path)

```
