```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
plt.rcParams['font.family'] = 'Arial'
import WormTool
```

```python
data = np.load('../neuron_ablations/curvature/N2.npy',allow_pickle=True)
i = 30
time = data[i,1][:,0] - data[i,1][0,0]
kt = -data[i,0][:15,:].mean(axis=0)
body = data[i,0][45:60,:].mean(axis=0)
lf,dn,_ = WormTool.timeseries.find_lf_denoised(kt)
hf = dn-lf
fig,ax = plt.subplots(1,1,dpi=300,figsize=(7,4))
ax.plot(time,kt,'g',label='head')
ax.plot(time,body,'k',label='body') 
plt.legend(loc='upper right')
#remove top and right spines
_ = [ax.spines[spine].set_visible(False) for spine in ['top','right']]
#set ticks inwards
ax.set_ylabel('Curvature ($\kappa$L)')
ax.set_xlabel('Time (s)')
plt.savefig('curvature.pdf')
plt.show()
```
